```python
def age_groups(row):
    if row["age"] < 30:
        return "20 - 30"
    elif row["age"] < 40:
        return "30 - 40"
    elif row["age"] < 50:
        return "40 - 50"
    elif row["age"] < 60:
        return "50 - 60"
    else:
        return "60 - 70"
```

```python
df ["age_range"]=df.apply(age_groups,axis =1)
```

Feature Engineering

```python
X = df[["age","annual Salary","net worth"]]
y= df[["car purchase amount"]]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train),len(X_test)
```

```python
len(y_train), len(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
import joblib
joblib.dump(scaler,"scaler.pkl")
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
def model_result(predictions):
    print("Mean Absolute error in this model is {}".format(mean_absolute_error(y_test,predictions)))
    print("Mean Squared error in this model is {}".format(np.sqrt(mean_squared_error(y_test, predictions))))
```

ML model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
pred_lr=lr.predict(X_test) #predicting the test data
```

```python
model_result(pred_lr)
```

```python
from sklearn.svm import SVR
svrmodel = SVR()
```

```python
param_gridsvr = { "C" : [0.01, 0.1, 0.5], 
                  "kernel" :["linear","rbf","ploy"],
                  "degree":[2,3,4]  
}
```

```python
from sklearn.model_selection import GridSearchCV
gridsvr = GridSearchCV(svrmodel,param_gridsvr)
```

```python
gridsvr.fit(X_train, y_train.values.ravel())
```

```python
pred_gridsvr= gridsvr.predict(X_test)
```

```python
model_result(pred_gridsvr)
```

```python
gridsvr.best_params_
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr_model = RandomForestRegressor()
param_gridrfr = {
    'n_estimators': [10,11,12,13,14,15],
    'max_depth': [5, 10, 15]
  
}
```
